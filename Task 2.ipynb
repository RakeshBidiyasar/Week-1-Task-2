```python
import pandas as pd

file_path = "titanic.csv"
df = pd.read_csv(file_path)

df.loc[:, "Age"] = df["Age"].fillna(df["Age"].median())  
df.loc[:, "Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.loc[:, "Cabin"] = df["Cabin"].fillna("Unknown")  
df.dropna(subset=["Survived"], inplace=True)  

df = df[(df["Age"] >= 0) & (df["Fare"] >= 0)]  
df.loc[:, "Sex"] = df["Sex"].str.lower()  
df.loc[:, "Embarked"] = df["Embarked"].str.upper()
df.loc[:, "Age"] = df["Age"].astype(int)  
df.loc[:, "Fare"] = df["Fare"].astype(int)
df.loc[:, "Survived"] = df["Survived"].map({0: "No", 1: "Yes"})  
df.loc[:, "Pclass"] = pd.Categorical(df["Pclass"], categories=[3, 2, 1], ordered=True)  

df.drop_duplicates(inplace=True)

print(df.to_string())
print(df.head())
```
